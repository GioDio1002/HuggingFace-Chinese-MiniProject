# Data Collection

# Sentiment Analysis

```python
#Analyze each sentence and accumulate the probability and output of each sentence
def sentiment_analysis(text_ls):
    analyzer = SentimentAnalyzer(lang="en") #Define anlayzer
    text_ls = []
    for sent in tqdm(bot_kira):
      tmp = analyzer.predict(sent)
      probas = [] #Store all the probablities
      for p in tmp.probas:
          probas.append(tmp.probas[p])
      text_ls.append([tmp.output]+probas)
    text_ls = pd.DataFrame(text_ls,columns = ['output','NEU','POS','NEG'])
    return text_ls
```

```python
sentiment_bot_kira = sentiment_analysis(bot_kira)
sentiment_kira = sentiment_analysis(kira)
```

```python
display(sentiment_bot_kira.head())
display(sentiment_kira.head())
```

```python
#Count the number of positve, neutral and negative feelings
output_bot_kira = sentiment_bot_kira.value_counts('output')
output_kira = sentiment_kira.value_counts('output')
print(output_bot_kira)
output_kira
```

```python
#The average of all probabilites of feelings
probas_bot_kira = sentiment_bot_kira[['NEU','POS','NEG']].mean()
probas_kira = sentiment_kira[['NEU','POS','NEG']].mean()
display(probas_bot_kira)
probas_kira
```

The outputs of feelings and probability distribution are identical. Perhaps, Youshikage Kira knows the *universal standard* when introducing yourself or telling a story. Other sentiment anlyzer needs to be investigated for further studies. Frankly speaking, the analyzer generates different results each time, but the same outputs for two texts and thus the credibilty of the analyzer is relatively low.

# Emotion Analysis

```python
#Define analyzer
emotion_analyzer = EmotionAnalyzer(lang="en")
emotion_analyzer.predict('Give me a break.') # Verified by Jotaro Kujo, 'Yare Yare daze'
```

```python
#Analyze each sentence and accumulate the probability and output of each sentence
def emotion_analysis(text_ls):
    text_ls = []
    for sent in tqdm(bot_kira):
      tmp = emotion_analyzer.predict(sent)
      probas = [] #Store all the probablities
      category = []
      for p in tmp.probas:
          category.append(p)
          probas.append(tmp.probas[p])
      text_ls.append([tmp.output]+probas)
    text_ls = pd.DataFrame(text_ls,columns = ['output']+category)
    return text_ls
```

```python
emotion_bot_kira = emotion_analysis(bot_kira)
emotion_kira = emotion_analysis(kira)
display(emotion_bot_kira.head())
display(emotion_kira.head())
```

```python
def output_count(df,title):
  sns.set_theme(style="whitegrid")#Set color theme
  df = df.value_counts('output').reset_index()
  df.columns = ['emotion','counts']
  sns.barplot(x='emotion',y='counts',data=df)
  plt.title('Emotion Distribution'+title)
  plt.show()
```

```python
output_count(emotion_bot_kira,'Bot_Generated')
output_count(emotion_kira,'Original')
```

According to the statistics, we can see the original speech of Kira is mostly classified as others, followed by fear, disgust and sadness. If we consider the scene was about Kira's first time to face a stand user, the results did support the emotion of the script, *the fear of not living a quiet life*. If there are any tools to analyze the tone of voice actor, we may analyze better in the sentiment analysis.

```python
nltk.download('punkt')
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
tokenizer = WordPunctTokenizer()
```

```python
def decontracted(text_ls):
    phrase = ' '.join(text_ls)
    # specific
    phrase = re.sub(r"won['’‘`]t", "will not", phrase)
    phrase = re.sub(r"can['’‘`]t", "can not", phrase)
    phrase = re.sub(r"ain['’‘`]t", "am not", phrase)

    # general
    phrase = re.sub(r"n['’‘`]t", " not", phrase)
    phrase = re.sub(r"['’‘`]re", " are", phrase)
    phrase = re.sub(r"['’‘`]s", " is", phrase)
    phrase = re.sub(r"['’‘`]d", " would", phrase)
    phrase = re.sub(r"['’‘`]ll", " will", phrase)
    phrase = re.sub(r"['’‘`]t", " not", phrase)
    phrase = re.sub(r"['’‘`]ve", " have", phrase)
    phrase = re.sub(r"['’‘`]m", " am", phrase)

    #phrase = re.sub('([.;!?])', r' \1 ', phrase)
    phrase = re.sub(r'[^\w.?!;]', ' ', phrase)
    phrase = re.sub(' +', ' ', phrase)
    sentences = re.split('([.;!?] *)', phrase)
    sentences = [lemmatizer.lemmatize(i.lower()) for i in  sentences]
    return ' '.join(sentences)
```

```python
text_kira = decontracted(kira)
bot_text_kira = decontracted(bot_kira)
text_kira[:10]
```
